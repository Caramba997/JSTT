```python
import json
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
import pandas as pd
```

```python
file = open('../testpairs.json')
data = json.load(file)
```

```python
def reject_outliers(xj, yj, m = 2.):
    x = np.array(xj)
    y = np.array(yj)
    newx = []
    newy = []
    quantile = np.quantile(y, 0.99)
    non_outliers = 0
    for i in range(0, len(y) - 1):
        if y[i] < quantile:
            newx = np.insert(newx, non_outliers, x[i])
            newy = np.insert(newy, non_outliers, y[i])
            non_outliers += 1
    return newx, newy
```

```python
for metric, values in data.items():
  x = values['test']
  y = values['source']
  plt.scatter(x, y, s=2)
  plt.title(metric)
  plt.ylabel(metric.split('/')[1])
  plt.xlabel(metric.split('/')[0])
  plt.show()
  newx, newy = reject_outliers(x, y)
  plt.scatter(newx, newy, s=2)
  plt.title('No outliers')
  plt.ylabel(metric.split('/')[1])
  plt.xlabel(metric.split('/')[0])
  plt.show()
```
